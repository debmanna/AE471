Uploading the 'ufluc' ALL time series data

```python
df_ufluc_ALL = pd.read_pickle('ufluc_TS_Allpoints_Full.pkl.gz')
df_ufluc_ALL
```

```python
# Creating the Time Series for a 'VarName' in the Dataset !

#rand = np.random.randint(0,18000)
rand = 5620
print(rand)
VarName = "ufluc_"+str(rand)
df_ufluc_ALL[VarName]
```

```python
pts = IndexToLocation(DataFrames, rand)
pts
```

```python
# Plot of the Time Series
plt.plot(df_ufluc_ALL[VarName], marker='.', markersize=5, linestyle='-', label=f"(x = {pts[0]}, y = {pts[1]})")

plt.axhline(y=0, color = 'k')

plt.gcf().set_size_inches(12, 3) 
plt.gcf().set_dpi(150)  # Set DPI for high quality


plt.title(f"Time Series of {VarName}")
plt.ylabel(f"{VarName}")
plt.xlabel("Time Step ->")
plt.legend()

plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

# Plot Partial Autocorrelation Function (PACF)
fig, ax = plt.subplots(figsize=(12, 3))
plt.gcf().set_dpi(150)
plot_acf(df_ufluc_ALL[VarName], lags=100, ax=ax, markersize=2)

plt.title(f"Autocorrelation Functions (ACF) for '{VarName}'")
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.legend([f"(x = {pts[0]}, y = {pts[1]})"], loc='upper right')

plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
import matplotlib.pyplot as plt

# Plot Partial Autocorrelation Function (PACF)
fig, ax = plt.subplots(figsize=(12, 3))
plt.gcf().set_dpi(150)
plot_pacf(df_ufluc_ALL[VarName], lags=200, ax=ax, markersize=2)

plt.title(f"Partial Autocorrelation Functions (PACF) for '{VarName}'")
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.legend([f"(x = {pts[0]}, y = {pts[1]})"], loc='upper right')

plt.grid(True)
plt.show()
```
